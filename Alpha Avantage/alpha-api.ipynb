```python
url = 'https://www.alphavantage.co/query?function=CASH_FLOW&symbol=IBM&apikey=demo'
r = requests.get(url)
data_cf = r.json()
display(data_cf)

# extrair lista de relatórios anuais
annual_reports = data_cf['annualReports']

# criar um dataframe a partir dos dados
df = pd.json_normalize(annual_reports)

# adicionar a coluna 'symbol' ao DataFrame
df['symbol'] = data_cf['symbol']

# reorganizar as colunas
df = df[['symbol', 'fiscalDateEnding', 'reportedCurrency', 'operatingCashflow', 'paymentsForOperatingActivities', 'proceedsFromOperatingActivities']]
df = pd.DataFrame(df)
# exibir DataFrame
display(df)
```
